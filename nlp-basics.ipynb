### Noise Removal

```python
# Sample code to remove noisy words from a text

noise_list = ["is", "a", "this", "an","was", "at", "..."] 
def _remove_noise(input_text):
    words = input_text.split() 
    noise_free_words = [word for word in words if word not in noise_list] 
    noise_free_text = " ".join(noise_free_words) 
    return noise_free_text

_remove_noise("Python is an interpreted high-level general-purpose programming language")
```

### Removing Regex Pattern 

```python
import re 

def _remove_regex(input_text, regex_pattern):
    urls = re.finditer(regex_pattern, input_text) 
    for i in urls: 
        input_text = re.sub(i.group().strip(), '', input_text)
    return input_text

regex_pattern = "#[\w]*"  

_remove_regex("#Python is an interpreted high-level general-purpose programming language ", regex_pattern)
```

### Lemmatization

```python
from nltk.stem.wordnet import WordNetLemmatizer 
lem = WordNetLemmatizer()

from nltk.stem.porter import PorterStemmer 
stem = PorterStemmer()

word = "playing" 
lem.lemmatize(word, "v")

stem.stem(word)
```

### Stemming

```python
pts = PorterStemmer()
 
# input for stemming
words = ["consult", "consultant", "consulting", "consultantative", "consultants"]
 
for w in words:
    print(w, " : ", pts.stem(w))
```
